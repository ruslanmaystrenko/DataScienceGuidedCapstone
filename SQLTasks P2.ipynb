```python
# Q10

database = "sqlite_db_pythonsqlite.db"
conn = create_connection(database)
cur = conn.cursor()

query10 = """
SELECT f.name AS facility_name, 
       SUM(CASE 
           WHEN b.memid = 0 THEN b.slots * f.guestcost 
           ELSE b.slots * f.membercost 
       END) AS total_revenue
FROM Bookings AS b
JOIN Facilities AS f ON b.facid = f.facid
GROUP BY f.name
HAVING total_revenue < 1000
ORDER BY total_revenue;
"""

cur.execute(query10)
rows = cur.fetchall()
print("Q10: Facilities with total revenue less than 1000")
print()
for row in rows:
    print(row)
```

```python
# Q11

cur = conn.cursor()

query11 = """
SELECT 
    m.firstname || ' ' || m.surname AS member_name,
    CASE 
        WHEN r.firstname IS NULL THEN 'None'
        ELSE r.firstname || ' ' || r.surname 
    END AS recommender_name
FROM Members AS m
LEFT JOIN Members AS r ON m.recommendedby = r.memid
WHERE m.memid != 0
ORDER BY m.surname, m.firstname;
"""

cur.execute(query11)
rows = cur.fetchall()
print("Q11: Member Name | Recommended By")
print()
for row in rows:
    print(f"{row[0]} | {row[1]}")
```

```python
# Q12
cur = conn.cursor()

query12 = """
SELECT f.name AS facility_name, m.surname, m.firstname, SUM(b.slots) AS usage_slots
FROM Bookings AS b
JOIN Facilities AS f ON b.facid = f.facid
JOIN Members AS m ON b.memid = m.memid
WHERE b.memid != 0
GROUP BY f.name, m.surname, m.firstname
ORDER BY f.name, m.surname, m.firstname;
"""

cur.execute(query12)
rows = cur.fetchall()
print("Q12: Facility Name | Surname | Firstname | Usage")
print()
for row in rows:
    print(f"{row[0]} | {row[1]} | {row[2]} | {row[3]}")
```

```python
# Q13

cur = conn.cursor()

query13 = """
SELECT f.name AS facility_name, 
       strftime('%Y-%m', b.starttime) AS month, 
       SUM(b.slots) AS total_slots
FROM Bookings AS b
JOIN Facilities AS f ON b.facid = f.facid
WHERE b.memid != 0
GROUP BY f.name, month
ORDER BY f.name, month;
"""

cur.execute(query13)
rows = cur.fetchall()
print("Q13: Facility Name | Month | Total Usage")
print()
for row in rows:
    print(f"{row[0]} | {row[1]} | {row[2]}")
```
